```python
# Necessary libraries
import numpy as np
import random
import matplotlib.pyplot as plt
import scipy.stats as scpy

# A Stats package that uses scipy as a basis - necessary to test for exponential distribution 
import statsmodels.api as sm
```

### Part1: Conjecture 2 - testing that all results are even

```python
# A Boolean to keep going until an odd end point is found
all_even = True

# n = 5 is the first we want to inestigate, so start here
n = 5

# keep going until od end point found
while all_even == True:

# 100 felt like a good number of repeats for each n, as its large enuogh to get statsitcally sufficent data but wont take impractical time to finish
    for i in range(0,100):

# initial hat
        hat = np.arange(1,n,1)
        hat = hat.tolist()

# playing the game by randomly selecting entries in hat and returning their postive difference
        while len(hat) > 1:
            pick1 = random.choice(hat)
            hat.remove(pick1)
            pick2 = random.choice(hat)
            hat.remove(pick2)
            diff = max(pick1, pick2) - min(pick1, pick2)
            hat.append(diff)

# checking whether the end point is divisible by 2 and if not the game will repeat
        final = hat[0]
        if (final % 2) != 0:
            all_even = False
        else:
            all_even = True
            i = i +1
            n = n +1

# retrurning the first end point that is odd and the n it occurs at
print("For n =", n, "not all results are even")
```

### Part 2: Code to take any n, k as input and return the hat game of size n repeated k times with end results in array


```python
final = []
n = 6
k = 100
for i in range(0,k):
    
    hat = np.arange(1,n,1)
    hat = hat.tolist()
    
    while len(hat) > 1:
        pick1 = random.choice(hat)
        hat.remove(pick1)
        pick2 = random.choice(hat)
        hat.remove(pick2)
        diff = max(pick1, pick2) - min(pick1, pick2)
        hat.append(diff)
        
    final.append(hat[0])
    i = i + 1
    
print(final)
```

### Part 3: Playing the hat game k times for any n again but instead returning the histogram and list (+count) of unique end points

```python
final = []
final_unique = []

n = 2024
k = 500
for i in range(0,k):
    
    hat = np.arange(1,n,1)
    hat = hat.tolist()
    
    while len(hat) > 1:
        pick1 = random.choice(hat)
        hat.remove(pick1)
        pick2 = random.choice(hat)
        hat.remove(pick2)
        diff = max(pick1, pick2) - min(pick1, pick2)
        hat.append(diff)
        
    end_point = hat[0]
    final.append(end_point)
    if end_point in final_unique:
        i = i + 1
    else:
        final_unique.append(end_point)
        i = i + 1

final_unique.sort()
print(final_unique)
fig, ax = plt.subplots()
ax.hist(final, color = '#21deaf')
plt.xlabel("final value")
plt.ylabel("count")
plt.title("Histogram of end values")
```

### Part 4: Same as above but outputting the density histograms with an exponential fit line overalyed (one with default bins, the other with bins containing at most one end point)

```python
final = []

n = 2024
k = 1000
for i in range(0,k):
    
    hat = np.arange(1,n,1)
    hat = hat.tolist()
    
    while len(hat) > 1:
        pick1 = random.choice(hat)
        hat.remove(pick1)
        pick2 = random.choice(hat)
        hat.remove(pick2)
        diff = max(pick1, pick2) - min(pick1, pick2)
        hat.append(diff)
        
    end_point = hat[0]
    final.append(end_point)
    if end_point in final_unique:
        i = i + 1
    else:
        final_unique.append(end_point)
        i = i + 1

lily = sm.stats.diagnostic.lilliefors(final, dist = 'exp', pvalmethod = 'table')
print(lily)

x = np.linspace(min(final), max(final),5000)
exp_fit = scpy.expon.pdf(x, scale = np.mean(final))


fix, ax = plt.subplots()
ax.hist(final, color = '#21deaf', label = "test sample of end value", density = True)
ax.plot(x, exp_fit, color = 'red', label = "exponential distribution fit line")
plt.xlabel("final value")
plt.ylabel("density (experimental and expected)")
plt.title("density histogram (default bins)")
plt.legend()

fix, ax = plt.subplots()
ax.hist(final, bins = int(max(final)/2), color = '#21deaf', label = "test sample of end value", density = True)
ax.plot(x, exp_fit, color = 'red', label = "exponential distribution fit line")
plt.xlabel("final value")
plt.ylabel("density (experimental and expected)")
plt.title("density histogram (single element bins)")
plt.legend()
```
